# Rational Numbers Lecture

```python
from fractions import Fraction
import math
```

```python
# Get some help on the fraction class
help(Fraction)
```

## 1. Constructor


```python
# Fraction defaults to numerator = 0, and denominator = 1
Fraction()
```

```python
# use an int to create it
x = Fraction(numerator=1,denominator=2)
print(x)
# use a float to create a fraction
x = Fraction(0.125)
print(x)
# use a string to set a floating num
x = Fraction('0.125')
print(x)
# use a string representation of a fraction
x = Fraction('22/7')
print(x)
```

## Attributes

```python
# We can access numerator and denominator
print(Fraction(1, -2).numerator)
print(Fraction(1, -2).denominator)
# See how it defaults the negative to the numerator
```

## Operations


```python
# Sum
x = Fraction(1, 2) + Fraction(1, 2)
print(x)
# Sub
x = Fraction(1, 2) - Fraction(1, 2)
print(x)
# Mult
x = Fraction(1, 2) * Fraction(1, 2)
print(x)
# Div
x = Fraction(1, 2) / Fraction(1, 2)
print(x)
```

## All Rationals

```python
# All nunmbers stored in a computer are finite which means they can be represented as rational numbers
x = Fraction(math.pi)
print(f"internal representation of pi as a fraction {x}")
print(f"{x} = {float(x)}")
```

```python
# Some internal representations make for long rationals which can be fixed by limiting the denominator
x = Fraction(0.125)
y = Fraction(0.3)

print(f"Representacion de 0.2 = {x}")
print(f"Representacion de 0.3 = {y}")
# 0.3 is not stored as 0.3 internally which is why the fraction is so long
print(f"0.3 in reality is {0.3: .25f} internally")
```

```python
# To fix this we can limin the denominator
x = Fraction(0.3)
x = x.limit_denominator(10)

print(f"With the limit_denominator to 10 we have 0.3 = {x}")
```
