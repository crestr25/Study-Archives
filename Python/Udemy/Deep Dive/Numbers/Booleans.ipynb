## Subclass of Int

```python
# Bool inherits from int
print(f"is bool a subclass of Int => {issubclass(bool, int)}")

# since they are singleton objects they don't change address
print(f"the address of True => {hex(id(True))} and the int value is => {int(True)}")
print(f"the address of False => {hex(id(False))} and the int value is => {int(False)}")
```

```python
# since they are singleton we can use the is operator with them
print(f"3 < 2 id is {hex(id(3 < 2))} which is the same as False {hex(id(False))}")

# since they have an int value we can say
print(f"1 + True = {1 + True}")
print(f"100 * False = {100 * False}")
print(f"True > False = {True > False}")
print(f"-True = {-True}")
```

# Constructor

```python
# The bool constructor will take the truthsy value of the object being passed
# it will evaluate the truth value of said object
print(f"bool(5) => {bool(5)}")
print(f"bool(0) => {bool(0)}")
print(f"bool(-1) => {bool(-1)}")
# What is actually doing is evaluating the __bool__ or the __len__
print(f"bool(5.__bool__()) => {bool((5).__bool__())}")
print(f"bool(list().__len__()) => {bool((list()).__len__())}")
```

```python
# Python actually implements the truthinness of the object for an statement
a = [1, 2, 3]
if a is not None and len(a) > 0:
  print("i exist")
else:
  print('Nothing here')

a = [1]
if a:
  print("i exist")
else:
  print('Nothing here')

a = [1, 2, 3]
if a.__len__(): # bool(a)
  print("i exist")
else:
  print('Nothing here')
```
