# Complex Numbers

```python
# When dealing with complex we use complex math
import cmath
```

```python
# constructor for complex numbers
a = complex(1,2)
b = 1 + 2j
print(f"I was constructed with the complex method: {a}")
print(f"I was constructed with the declarative method: {b}")
print(f"a == b => {a == b}")
```

```python
# For complex numbers x and y (x + yj) are stored as float
print(f"the type of the x is {type(a.real)}")
print(f"the type of the y is {type(a.imag)}")
```
